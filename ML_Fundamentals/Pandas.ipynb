**Transform data using apply()**

```python
customers_data=pd.read_csv('customer_profiles-1-1714027410.csv')
customers_data
```

```python
def increment(x):
    return x+5

customers_data['Age']=customers_data['Age'].apply(increment)
```

```python
customers_data["Age"]
```

```python
customers_data["Age"]=customers_data["Age"].apply(lambda x:x-5)
```

```python
customers_data["Age"]
```

```python
customers_data["Age category"]=customers_data["Age"].apply(lambda x: "Very young" if x < 25 else "Not so young")
```

```python
customers_data.head
```

```python
customers_data.insert(2,"Rating",5)
```

```python
customers_data.head
```

```python
customers_data.drop(columns="Rating")
```

```python
del customers_data["Age category"]
```

**Merge,Join & Concatenate Dataframes**

```python
df1=pd.DataFrame({
    "id":[1,2,3,4],
"name":["rakesh","virat","kohli","kumar"]
})
df2=pd.DataFrame({
    "id":[1,6,7,4],
"name":["v","vault","veena","k"]
})
inner_result=pd.merge(df1,df2,how="inner",on='id')
inner_result
```

```python
df1=pd.DataFrame({
    "id":[1,2,3,4],
"name":["rakesh","virat","kohli","kumar"]
})
df2=pd.DataFrame({
    "id":[1,2,3,4],
"name":["v","vault","veena","k"]
})
outer_result=pd.merge(df1,df2,how="outer",on='id')
outer_result
```

**Concatenation**

```python
df1=pd.DataFrame({
    "id":[1,2,3,4],
"name":["rakesh","virat","kohli","kumar"]
})
df2=pd.DataFrame({
    "id":[1,2,3,4],
"name":["v","vault","veena","k"]
})
result=pd.concat([df1,df2],axis=0)
```

```python
result
```
